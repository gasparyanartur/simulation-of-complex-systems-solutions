```python
import numpy as np
import matplotlib.pyplot as plt
import math
rng = np.random.default_rng()
```

## Exercise 10.1

```python
n_steps = 1000000
time_step = 0.02
gaussian_seq = rng.normal(size=(n_steps,))
 
v0 = 2
tau = 2

def simulate_motion(gaussian_seq):
    x = np.zeros((n_steps, 3))
    
    v = v0
    fluct = math.sqrt(2/tau) * gaussian_seq
    
    for i in range(n_steps):
        theta = x[i-1, 2]
        x[i, 0] = x[i-1, 0] + v*math.cos(theta)*time_step
        x[i, 1] = x[i-1, 1] + v*math.sin(theta)*time_step
        x[i, 2] = x[i-1, 2] + fluct[i]*time_step
        
    return x


x = simulate_motion(gaussian_seq)
print(x)
plt.plot(x[:, 0], x[:, 1])
```
